## As Equações Da Estática

Se colocarmos $v = 0, \frac{\partial}{\partial t} = 0$ e $\sigma_{ij} = -P\delta_{ij}$ nas leis do movimento de fluidos. Nós achamos a mecânica e o equilíbrio térmico de fluidos estáticos governados por:

$$ \begin{align*} -\nabla P +f & =0 \\ \nabla \cdot (\mathcal{X} \nabla T) + \mathcal{Q} & =0\end{align*}$$

Onde $f$ é um campo de força volumétrica aplicada e $\mathcal{Q}$ uma densidade de fonte de calor. Nós
observe imediatamente que se $f$ for zero, a pressão é uniforme.

O primeiro resultado importante das equações acima é que existe uma solução estática
se, e somente se, a força externa pode ser derivada de um potencial. Assim, podemos definir $f = - \nabla \phi_{ext}$ e resolver a pressão

$$ P + \phi_{ext} = Cst$$

Esta solução mostra que os isóbaros são idênticos às superfícies equipotenciais. Nós agora
saibemos que se $f$ não for o gradiente de um potencial, não existe solução estática. O fluido
flui. A equação $\nabla \cdot (\mathcal{X} \nabla T) + \mathcal{Q} =0$  nos dá o campo de temperatura. Se a condutividade térmica for constante ou uma função suave das coordenadas espaciais, esta equação tem uma solução. Na maioria dos casos, $f$ é proporcional à densidade $\rho$.

REFERÊNCIA: RIEUTORD, Michel. Fluid Dynamics An Introduction. Masson, France: Springer, 1997.
