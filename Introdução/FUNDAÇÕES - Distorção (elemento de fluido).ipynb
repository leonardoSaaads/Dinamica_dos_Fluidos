# FUNDAMENTOS - FLUIDOS

## Distorção de um elemento de fluido

Um aspecto importante do movimento das partículas de fluido é seu movimento adequado. De fato, mencionamos acima que, embora de tamanho pequeno, as partículas de fluido possuem uma superfície e um volume. Assim, eles podem ser distorcidos por um fluxo não uniforme. Para visualizar este efeito, é conveniente considerar uma partícula de fluido de paralelepípedo forma que podemos caracterizar por um vetor $\xi$. Se expressarmos o variação de $\xi$ com o tempo, encontramos:

![](https://raw.githubusercontent.com/leonardoSaaads/Dinamica_dos_Fluidos/main/Imagens/livro_teste.png)

$$ \xi + \delta \xi = \xi + v(x+\xi)\delta t - v(x)\delta t$$

Supondo que $\xi$ seja muito pequeno em comparação com x, temos

$$ v(x + \xi) = v(x) + (\xi \cdot \nabla)v + O(\xi^2) $$

Na primeira ordem:

$$ \delta \xi = (\xi \cdot \nabla)v \delta t$$

Usando notações indiciais

$$ \delta \xi_i = \xi_j \partial_j v_i \delta t$$

Nesta expressão descobrimos o tensor “gradiente de velocidade” cujos componentes são $\partial_i v_i$. Como qualquer tensor de segunda ordem, ele pode ser decomposto em seu simétrico e anti-simétrico:

$$ \partial_i v_j = \frac{1}{2} (\partial_i v_j + \partial_j v_i) + \frac{1}{2} (\partial_i v_j + \partial_j v_i) =s_{ij} + a_{ij}$$

Essas duas partes desempenham papéis muito diferentes. Vamos primeiro nos concentrar no antissimétrico.
Notamos que ele é representado por apenas três componentes $(a_{12}; a_{23}; a_{31})$. Na verdade, esses componentes estão relacionados à curvatura do campo de velocidade. De fato,

$$ a_{ij} = \frac{1}{2} (\partial_i v_j + \partial_j v_i) = \frac{1}{2} (\delta_{ik} \delta_{jl} + \delta_{il} \delta_{jk})\partial_k v_l = \frac{1}{2} \epsilon_{ijm} \epsilon_{mkl} \partial_k v_l $$

Onde introduzimos o delta de Kronecker $\delta_{ij}$ e o tensor completamente anti-simétrico $\epsilon_{ijk}$ (seus componentes são apenas $\pm$1 ou 0). Portanto,

$$ a_{ij} = \frac{1}{2} \epsilon_{ijm}(\nabla \times v)_m $$

essa expressão mostra que as três componentes $(a_{12}; a_{23}; a_{31})$ são apenas as componentes da curvatura de v multiplicadas por um fator de 1/2, uma vez que $\epsilon_{ijk}$ é zero ou unidade. A expressão é útil para compreender o significado físico de \[a\] e, conseqüentemente, da curvatura. Podemos descobri-la calculando a variação de $\xi$ associado a \[a\] apenas. Portanto:

$$ \delta \xi_i = \xi_j \partial_j v_i \delta t = \frac{1}{2} \epsilon_{jim}\xi_j (\nabla \times v)_m \delta t = \left( \frac{1}{2} (\nabla \times v) \times \xi\right)_i \delta t$$

Esta expressão mostra que a variação de um elemento fluido associado a \[a\] é apenas uma rotação de corpo sólido na velocidade angular

$$ \Omega = \frac{1}{2} (\nabla \times v) $$

Este resultado nos dá a interpretação física do vetor $\nabla \times v$, também chamado de vorticidade e esclarece a onda de formulação para o operador $\nabla \times$.
    Uma vez que a rotação de um corpo sólido não distorce nosso elemento fluido paralelepipédico, sua deformação deve estar contida na parte restante do tensor gradiente de velocidade, ou seja, \[s\]. Essa parte simétrica do tensor de gradiente de velocidade é chamada de tensor de taxa de deformação. O efeito de distorção de \[s\] em um elemento fluido pode ser explicado por considerando as variações do comprimento do vetor. De fato,

$$ \delta(\xi^2) = 2\xi_i \xi_i = 2\xi_i \xi_j \partial_i v_j \delta t = 2\xi_i \xi_j s_{ji} \delta t$$

onde usamos o fato de que $\xi_i \xi_j a_{ij} = 0$, uma vez que $a_{ij}$ é antissimétrico e $\xi_i \xi_j$ é simétrico. Assim, apenas \[s\] contribui para a variação do comprimento de $\xi$ e, portanto, à distorção da partícula de fluido. A fim de fundamentar mais completamente o efeitos da deformação, é útil expressar as variações de $\xi$  em uma base onde \[s\] é diagonal. Essa base sempre existe porque \[s\] é simétrica. Nesta nova base a variação de associado com \[s\] é

$$ \delta \xi_i = \xi_j s_{ji} \delta t = \xi_i s_{ii} \delta t$$

onde nenhuma soma é assumida nos índices na última expressão. Esta equação mostra que o elemento de fluido é alongado na direção i quando $s_{ii}> 0$ ou desenrolado quando $s_{ii} <0$. Então, é fácil calcular a sua variação de volume no início pedido devido ao campo de velocidade:

$$ \delta V = (\xi_1 + \delta \xi_1)(\xi_2 + \delta \xi_2)(\xi_3 + \delta\xi_3) - \xi_1 \xi_2 \xi_3 = \xi_1 \xi_2 \xi_3 \left( \frac{\delta \xi_1}{\xi_1} + \frac{\delta \xi_2}{\xi_2} +\frac{\delta \xi_2}{\xi_2}\right) + O(\delta \xi^2) =  \xi_1 \xi_2 \xi_3 (s_{11} + s_{22} + s_{33}) + O(\delta \xi^2)$$

Mas $s_{11} + s_{22} + s_{33} = T(s) = \nabla \cdot v$, então:

$$ \frac{\delta V}{V} = (\nabla \cdot v)\delta t $$

Este resultado mostra que a dilatação ou contração dos elementos fluidos está associada com o traço do tensor de taxa de deformação, que também é a divergência do campo de velocidade. Assim, elementos de expansão aparecem em fluxos com divergência positiva enquanto os elementos de contração estão onde a divergência é negativa. Observe que este resultado é independente da base escolhida, pois o traço de um tensor é invariante nas mudanças de base.
