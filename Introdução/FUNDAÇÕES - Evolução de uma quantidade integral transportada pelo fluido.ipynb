# FUNDAMENTOS - FLUIDOS

## Evolução de uma quantidade integral transportada pelo fluido.

A seguir, conheceremos a evolução temporal de uma quantidade (massa, momento, energia) associada a um domínio de fluido D (volume, superfície, linha). No caso de os pontos do domínio do fluido têm uma velocidade igual à do fluido a cada instantâneo, o domínio é chamado de domínio material. É sempre feito do mesmo fluido partículas. Em outras palavras, esse domínio é uma espécie de partícula macroscópica.

   Mais tarde, precisaremos da expressão da derivada material de uma quantidade $f$ integrado sobre um domínio de material $V(t)$. Então:


$$ \displaystyle \frac{d}{dt} \int_{V(t)} f(r, t) dV $$

em função das derivadas locais de $f$. Para isso, nós escrevemos

$$ \displaystyle \frac{d}{dt} \int_{V(t)} f(r, t) dV  = \frac{\int_{V(t + dt)} f(r + dr, t + dt) dV' -  \int_{V(t)} f(r, t) dV}{dt}$$

Desenvolvemos a primeira integral de primeira ordem e, uma vez que $dr = v \cdot dt$, obtemos:

$$ f(r + dr, t + dt) = f(r, t) + \frac{Df}{Dt}dt $$

No entanto, o volume $V(t + dt)$ é ligeiramente diferente do volume $V(t)$ representa como o campo de velocidade o distorce. Para levar em conta essa distorção, ainda devemos integrar mais de $V(t)$ mas com um volume elementar distorcido $dV'$.

$$ dV' = dV + dV\nabla \cdot v dt = dV(1 + \nabla \cdot v dt) $$

Usando essa expressão, podemos notar que:

$$ \displaystyle \frac{d}{dt} \oint_{C(t)} A(r,t) \cdot dl $$

Como acima, precisamos considerar as variações do elemento fluido $dl$, que é modificado no tempo $t+dt $, assim como $\xi$ em $\delta \xi = (\xi\ \cdot \nabla)v \delta t$. Então: 

$$ dl' = dl + (dl \cdot \nabla)v dt $$

Logo:

$$ \displaystyle \frac{d}{dt} \oint_{C(t)} A(r,t) \cdot dl =  \displaystyle \oint_{C(t)} \left( \frac{DA}{Dt}\cdot dl + A\cdot (dl \cdot \nabla)v\right) $$

O que é mais útil se usarmos índices:

$$\displaystyle \frac{d}{dt} \oint_{C(t)} A(r,t) \cdot dl = \oint_{C(t)} \left( \frac{DA_i}{Dt}\cdot dl + A_j\partial_i v_j\right) dl_i$$

A Expressão pode ser simplificada para:

$$ A_j \partial_i v_j = \partial_i(A_j v_j) - v_j \partial_i A_j \,\, , \, \, \partial_j A_i- \partial_i A_j  = \epsilon_{jik}(\nabla \times A)_k $$

Finalizando em:

$$ \displaystyle \frac{d}{dt} \oint_{C(t)} A(r,t) \cdot dl = \oint_{C(t)} \left( \frac{\partial A_i}{\partial t} - (v \times (\nabla \times A))_i\right) dl_i $$

REFERÊNCIA: RIEUTORD, Michel. Fluid Dynamics An Introduction. Masson, France: Springer, 1997.
