# FUNDAMENTOS - FLUIDOS

## Derivada Material

Muitas vezes, vamos considerar a evolução do tempo de alguma quantidade, como $\phi$, que é anexado a uma partícula fluida que se segue.Para expressar as variações de $\phi$, introduzimos a derivada:

$$ \frac{D\phi}{Dt}$$

também chamada de derivada material ou derivada lagrangiana. $\phi$ pode ser qualquer quantidade (escalar, vetor, tensor) dependendo das coordenadas espaciais e do tempo $\phi = \phi(x,y,z,t)$. Quando nos determinamos $\phi$ ao movimento de uma partícula fluida, as coordenadas x, y, z são funções do tempo.

$$\phi = \phi(x(t),y(t),z(t),t)$$

onde $x(t), y(t), z(t)$ representam a trajetória da partícula fluida. Agora, o total variação $\phi$ de é:

$$ \frac{D\phi}{Dt} = \frac{\partial \phi}{\partial x} \cdot \frac{dx}{dt} + \frac{\partial \phi}{\partial y} \cdot \frac{dy}{dt} + \frac{\partial \phi}{\partial z} \cdot \frac{dz}{dt} + \frac{\partial \phi}{\partial t} \tag{via regra da cadeia}$$

Isso resulta em:

$$ \frac{D\phi}{Dt} = \frac{\partial \phi}{\partial x}\dot v_x + \frac{\partial \phi}{\partial y}\dot v_y + \frac{\partial \phi}{\partial z}\dot v_x + \frac{\partial \phi}{\partial t}$$
ou
$$ \frac{D\phi}{Dt} = \frac{\partial \phi}{\partial t} + (v \cdot \nabla)\phi $$

O operador D/Dt simboliza, portanto, uma variação ao longo de uma curva, ou seja, ao longo da trajetória de um elemento fluido. Veremos que essa quantidade aparece com muita frequência nas equações que regem os fluxos de fluidos. O termo $(v \cdot \nabla)$ é chamado de termo de advecção e representa o transporte da quantidade $\phi$ pelo campo de velocidade $v$. Em um fluxo constante, ele representa as variações de $\phi$ ao longo de uma linha.
Outra ilustração do papel desempenhado pelo termo de advecção é dada pelo caso onde $\phi$ é conservado por cada partícula. Assim:

$$ \frac{D\phi}{Dt} = 0 $$

Nesse caso, um observador medindo $\phi$ em um determinado ponto em função do tempo ver as variações de $\phi$ correspondentes às partículas que passam exibindo seu valor de $\phi$. Formalmente, isso significa:

$$ \frac{\partial \phi}{\partial t} = - (v \cdot \nabla)\phi $$

Este resultado mostra que a dilatação ou contração de elementos fluidos está associada com o traço do tensor taxa de tensão, que também é a divergência do campo de velocidade. Assim, elementos em expansão aparecem em fluxos com divergência positiva enquanto os elementos de contração são onde a divergência é negativa. Note que este resultado é independente da base escolhida como o traço de um tensor é invariante nas mudanças de base.

REFERÊNCIA: RIEUTORD, Michel. Fluid Dynamics An Introduction. Masson, France: Springer, 1997.
