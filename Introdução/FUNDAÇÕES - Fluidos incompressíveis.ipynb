# FUNDAMENTOS - FLUIDOS

## Fluidos Incompressíveis

Um modelo importante para a descrição dos fluxos de fluidos é o denominado fluido incompressível. Para tal fluido, a densidade é considerada constante. Essa suposição é muito popular porque simplifica muito as equações do movimento (e a física do fluido). Além disso, é uma boa aproximação para líquidos que são apenas ligeiramente compressível. Além disso, veremos mais tarde que mesmo os fluxos de gás podem ser modelados por tal fluido, desde que a velocidade seja pequena em comparação com a velocidade do som.

    Vemos que se o volume de um elemento fluido não varia, então:

$$ \nabla \cdot v = 0 $$

As partículas de fluido não se expandem nem se contraem. Esta é a principal restrição que deve ser
encontrado por um fluido cujas variações de densidade podem ser ignoradas. Devemos encontrar esta relação
novamente ao estudar as implicações da conservação de massa.

Na realidade, todos os fluidos são compressíveis, alguns mais que outros. A compressão de um fluido mede a alteração no volume de uma certa quantidade de líquido quando se submete a uma pressão exterior. Por exemplo, se é obstruída hermeticamente a saída de uma bomba de bicicleta e se empurra o êmbolo, vemos que podemos comprimir o ar que contém. Entretanto, se fazemos a mesma experiência com água em seu anterior, ou em dispositivo similar, como uma seringa, vemos que apenas podemos mover o êmbolo um mínimo, mais por deformação das próprias paredes da bomba ou seringa, porque a compressibilidade da água (e de qualquer líquido) é muito baixa.

Em termos matemáticos, isto significa que a densidade de tal fluido se supõe constante

$$ \rho = \rho_0 = const $$
