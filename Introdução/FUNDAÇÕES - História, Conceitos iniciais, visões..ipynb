## Breve Perspectiva Histórica.

A primeira etapa da Mecânica dos Fluidos foi certamente realizada por Arquimedes (-287, -212), que foi um matemático e físico na Antiguidade. Ele formulou um teorema agora conhecido que diz que um corpo imerso em um fluido suporta um impulso para cima igual ao peso do fluido deslocado. Este é o primeiro resultado da teoria do equilíbrio dos fluidos. O conhecimento não evoluiu muito até os trabalhos de Evangelista Torricelli (1608-1647) e Blaise Pascal (1623-1662), Torricelli fez uma famosa experiência ao colocar de cabeça para baixo um tubo cheio de mercúrio. O líquido desceu um pouco e deixou uma coluna de mercúrio com 76 cm de altura. Assim, foi demonstrada a existência de pressão atmosférica, o peso do ar e a existência de vácuo, muito discutidos na época. Pascal deu um relato completo de todos esses fenômenos em seu tratado L'équilibre des liqueurs, publicado em 1663. A estática dos fluidos estava quase instalada. A dinâmica dos fluidos começou com o trabalho de Leonhard Euler (1707-1783), que formulou pela primeira vez a equação do movimento de um fluido invíscido. Daniel Bernoulli (1700-1782) contribuiu para o estudo de tais fluidos com teoremas sobre conservação de energia que revelaram propriedades fundamentais dos fluxos de fluidos.

O próximo passo importante foi a formulação dos efeitos da viscosidade. Isso foi feito durante o século XIX com a obra de Henri Navier (1785-1836). Georges Stokes (1819-1904) e Jean-Louis-Marie Poiseuille (1799-1869). Notemos que Isaac Newton (1642-1727) já mostrava a existência da viscosidade com experimentos e deixou seu nome associado a uma espécie de fluidos (os mais comuns como o ar e a água) hoje conhecidos como fluidos newtonianos. A equação de Navier-Stokes, que controla a dinâmica de fluidos viscosos, foi formulada pela primeira vez por Navier no caso de um fluido com viscosidade constante.

Então, a mecânica dos fluidos tomou vários caminhos. Mencionaremos apenas os principais Os estudos de estabilidade de fluxos começaram com os trabalhos de Helmholtz (1868) e Lord Kelvin (1880). O transporte de calor foi estudado por W. Prout (1834), Rumford (1870). A. Oberbeck (1879), H. Benard (1900), J. Boussinesq (1903) e Lord Rayleigh(1916). A turbulência focou o interesse de O. Reynolds (1883), L. Prandtl (1920), A. Kolmogorov (1940), etc.

Atualmente, a dinâmica dos fluidos é ainda mais diversificada, mas alguns velhos problemas resistem:
a turbulência continua sendo um dos principais problemas não resolvidos da física clássica.

## Conceito de Fluido.

Os fluidos reúnem um grande número de formas de matéria que, tanto quanto líquidos e gases, podem ser caracterizados pela facilidade com que podem ser deformados. Este ponto de vista expressa a diferença entre o “estado sólido” e o “estado fluido”, que são ambos “estados mecânicos”. Comum experiência assume que tal diferença é óbvia, no entanto, veremos que um uma inspeção mais detalhada de alguma forma obscurece a diferença entre sólidos e fluidos. Na verdade, essa diferença está contida em uma lei, conhecida como lei reológica, que afirma como a matéria reage a (é deformada por) uma tensão (uma força por unidade de superfície).
De um modo geral, a matéria é considerada fluida se apenas as tensões internas (cisalhamento) dependem de gradientes de velocidade. Voltaremos a esses conceitos, mas vamos dar um exemplo simples para apreciar o que está por trás desta definição. Consideramos uma rolha flutuando na água em um container. A rolha está em A e queremos movê-la para B. Como a água é um fluido, a força que precisamos aplicar na cortiça depende apenas dos gradientes de velocidade. Pode ser feito tão pequeno quanto desejarmos; só precisamos mover a rolha mais devagar. Do ponto de vista da termodinâmica, podemos dizer que os estados A e B e não podem ser distinguidos: as variações de energia e entropia necessárias para fazer a passagem do sistema de um estado para o outro pode ser tornada extremamente pequena. Sobre pelo contrário, se fizéssemos um experimento semelhante de um sólido, algum trabalho teria sido necessário e a energia dos estados A e B seriam diferentes.

## Cinemática de Fluidos.

O primeiro passo necessário para entender o movimento dos fluidos é encontrar as ferramentas certas que nos permitirá descrever um fluxo de fluido. Este é o papel da cinemática fluida que estuda movimento fluido sem se preocupar com suas causas.

## O conceito de partícula fluida.

Muito frequentemente, usaremos o conceito de partícula fluida ou elemento fluido. Isto é um visão idealizada de um pedaço de fluido: é tão pequeno que as propriedades do fluido são uniformes lado de dentro. No entanto, é grande o suficiente para conter um grande número de átomos ou moléculas, o que nos permite supor que o fluido está localmente em termodinâmica equilíbrio (portanto, a temperatura é definida). Essa partícula não é uma massa pontual: possui uma superfície que autoriza forças de contato com outras partículas.

###  *A Visão Lagrangiana*

Uma primeira maneira de descrever o movimento do fluido é dar as trajetórias de todos os partículas de fluido. Essa descrição é a Lagrangiana. Pode ser resumido pelo conjunto de trajetórias de todas as partículas de fluido que estavam no domínio D_{t0}
em t=t0. Matematicamente, é o conjunto.

$$ D_{t} = \{x(t,x_{a}) | x_{a} \in D_{t0}, t ≥ t_{0} \} $$


Tal descrição é usada em alguns estudos específicos onde fornece simplificações. No entanto, de um modo geral, seu uso não é muito popular devido a algumas dificuldades intrínsecas na expressão de estresses. Encaminhamos o leitor ao final deste capítulo para uma apresentação mais detalhada.

###  *A Visão Euleriana.*

A maneira mais natural de representar o movimento de um fluido é certamente a intuitiva usado para descrever o fluxo de água em um rio. Em tal caso, nós diria que o fluxo é rápido em um lugar em um dado momento e dificilmente perceptível em algum outro lugar e em outra ocasião. Assim, descreve-se o campo de velocidade como uma função de espaço e tempo. Na verdade, se a função vetorial.

$$ v(x, y, z, t) $$

é conhecido, então tudo é conhecido sobre o fluxo de fluidos. O uso do campo de velocidade em função da posição e do tempo dá a descrição euleriana do fluxo. Isso é a maneira mais popular de fazer e vamos usá-lo quase o tempo todo a partir de agora.
