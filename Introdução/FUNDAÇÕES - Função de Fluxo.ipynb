## Função de Fluxo para Fluidos

Na parte de introdução, é possível ver que o fluxo de um fluido incompressível deve atender à restrição $\nabla \cdot v = 0$. Vamos agora considerar um outro caso simplificado quando o fluxo do fluido é bidimensional. Se x e y são coordenadas cartesianas neste plano, $\nabla \cdot v = 0$ nos leva a:

$$ \nabla \cdot v = 0 \implies \frac{\partial}{\partial x} v_x + \frac{\partial}{\partial y} v_y = 0$$

É fácil mostrar que a equação acima implica na existência de uma função $\psi(x,y,t)$ tal que

$$ v_x = \frac{\partial \psi}{\partial x} \, \, , \, \, v_y = \frac{\partial \psi}{\partial y} $$

___

### **Relembre-se: Método das Curvas Características**

Considere a seguinte Equação Diferencial Parcial de primeira ordem, linear, homogênea e de coeficientes constantes:


$$ \frac{\partial}{\partial x} u(x,y) + \frac{\partial}{\partial y} u(x,y) = 0 \, (a,b \in \mathbb{R}) \tag{1}$$

Se $u(x,y)$ é uma solução de (1), e $\vec V = <a,b>$, então: 

$$\frac{\partial u(x,y)}{\partial \vec V} = \nabla \vec u \cdot \vec V = < \frac{\partial}{\partial x} u(x,y), \frac{\partial}{\partial y} u(x,y) > \cdot < a, b > = \frac{\partial}{\partial x} u(x,y) + \frac{\partial}{\partial y} u(x,y) = 0 $$

Logo, $ \vec u(x,y)$ é constante ao longo das retas definidas por $\vec V$:

![](https://raw.githubusercontent.com/leonardoSaaads/Dinamica_dos_Fluidos/main/Imagens/vetor%20EDP.png)

Portanto, ao longo das linhas (características) $u(x,y) = f(c_1) \, , \, (c_1 \in \mathbb{R})$. Vamos agora determinar as retas características. para tanto, seja $\vec \gamma(t) = <x(t), y(t)>$. Por construção, $\vec \gamma'(t) = \vec V$. Ou seja:

$$ \vec \gamma '(t) = < \frac{dx}{dt}, \frac{dy}{dt} > = <a, b> \, , \, (a,b ≠ 0) $$

Isso implica que

$$\left\{ \begin{align*} \frac{dx}{dt} & =a \\ \frac{dy}{dt} & =b \end{align*} \right. \implies \boxed{\frac{dx}{a} = \frac{dy}{b}} \tag{2}$$

Resolvendo-se (2), temos que:

$$\displaystyle b\int dx  = a\int dy \implies bx = ay + c \implies c_1 = bx - ay \, , \, (c_1 \in \mathbb{R})$$

Finalizando-se, a equação pode ser escrita como:

$$ u(x, y) = f(c_1) = f(bx - ay)$$

Chamada de função fluxo. A expressão de um fluxo bidimensional é, portanto, contido no de um campo escalar único, ou seja, a função de fluxo. Esta redação explicita o fato de que isolinhas da função de fluxo são sempre tangenciais ao campo de velocidade e, portanto, trace as linhas de campo de velocidade ou as linhas de corrente. Na verdade, ao longo de uma linha:

$$ d\psi = 0 \iff \frac{\partial \psi}{\partial x}dx + \frac{\partial \psi}{\partial y}dy = 0 \iff \begin{vmatrix} v_x & dx \\ v_y & dy \end{vmatrix} = 0 \iff v \,  || \,   {dl} $$

Quando o fluxo é tridimensional, essa ideia pode ser generalizada, mas dois fluxos funções são necessárias. Isso é facilmente compreendido se lembrarmos que em três dimensões, o campo de velocidade tem três componentes que são restringidos por um equação. Conseqüentemente, fica-se com duas quantidades independentes. Sem perda de generalidade, pode-se escrever

$$ v = \nabla \times (\chi a) + \nabla \times \nabla \times (\psi a) $$

Nesta expressão, a escolha de a não é imposta. Na geometria esférica, geralmente se escolhe o vetor radial r. O primeiro termo é então chamado de campo de velocidade toroidal - porque as linhas de campo estão em um toro. O segundo termo é o campo poloidal; seu campo as linhas geralmente não estão confinadas em uma superfície, exceto no caso de simetria axial, onde eles podem ser desenhados em um plano meridiano $\varphi = Const$.
