# FUDAMENTOS - FLUIDOS

## Termodinâmica - Gás Ideal

A expressão da energia interna de um gás ideal em função das variáveis externas é:


$$ e = e_0 \left( \frac{\rho}{\rho_0}\right)^{\gamma -1} e^{(s - s_0)/c_v}$$

Duas relações clássicas surgem dessa expressão:

$$ PV = nRT$$

e


$$ e = c_vT = \frac{1}{1-\gamma}\frac{P}{\rho} \, \, ,\, \, \gamma = \frac{c_p}{c_v} $$

Onde $R = k_b \mathcal{N}$ é a constante de gás ideal, ou a expressão macroscópica de
Constante de Boltzmann $k_b$ usando o número de Avogadro $\mathcal{N}$. $n$ é o número mol e $c_v$
(resp. $c_p$) é a capacidade de calor específico em volume constante (resp. pressão). A equação $PV = nRT$ pode ser escrita como

$$P = R_* \rho T$$

o que é mais conveniente na mecânica dos fluidos. $R_* = R/M$ onde M é a massa de
um mol de gás. Entalpia específicas nos leva a

$$ h = c_p T = \frac{\gamma}{\gamma -1}\frac{P}{\rho} $$

enquanto várias expressões de entropia podem ser derivadas de

$$Tds = de - \frac{P}{\rho^2}d\rho =dh + \frac{dP}{\rho}$$

por exemplo:

$$ \begin{align*}s &= c_v \ln(T/T_0) - R_*\ln(\rho/\rho_0) + s_0 \\
s &= c_p \ln(T/T_0) - R_*\ln(P/P_0) + s_0 \\
s &= c_v \ln(P/P_0) - c_p\ln(\rho/\rho_0) + s_0 \end{align*}$$

REFERÊNCIA: RIEUTORD, Michel. Fluid Dynamics An Introduction. Masson, France: Springer, 1997.
