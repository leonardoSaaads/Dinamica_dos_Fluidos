# FUNDAÇÕES - FLUIDOS

## Fluidos Barotrópicos

Um caso simétrico ao dos líquidos aparece quando a densidade é apenas uma função
de pressão


$$ \rho \equiv \rho(P) \, \, \, \, or \, \, \, \, P \equiv P(\rho)$$

Na maioria dos casos, esta não é uma equação de estado do fluido, mas um poço de aproximação verificado em certas circunstâncias.
    Dois exemplos são freqüentemente encontrados: os casos de um isotérmico ou de um isentrópico
gás ideal. Nesses casos, a pressão é apenas uma função da densidade

$$ P = k\rho \, \, \, \, ou \, \, \, \, P = K\rho^{\gamma}$$

Essa dependência surge quando as variações de temperatura ou entropia podem ser desprezadas.
No caso geral, tais relações não existem. Os fluxos são chamados de baroclínicos
porque isóbaras e isotérmicas são inclinadas uma em relação à outra. Barotropicidade
e a baroclinicidade pode ter consequências importantes sobre a natureza dos fluxos

REFERÊNCIA: RIEUTORD, Michel. Fluid Dynamics An Introduction. Masson, France: Springer, 1997.
