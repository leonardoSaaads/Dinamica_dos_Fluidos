# FUDAMENTOS - FLUIDOS

## Termodinâmica - Líquidos

Se nos concentrarmos em líquidos em condições usuais, a termodinâmica é simples porque os líquidos são pouco
compressíveis. Na maioria dos casos, as variações de densidade vêm principalmente das variações de temperatura. Um modelo simplificado consiste em reter apenas tal relação como

$$ \rho = \rho_0(1 - \alpha(T-T_0))$$

que é completado por e $e =cT$. $\alpha$ é o coeficiente de dilatação térmica e $c = c_v ≅ c_p$

REFERÊNCIA: RIEUTORD, Michel. Fluid Dynamics An Introduction. Masson, France: Springer, 1997.
