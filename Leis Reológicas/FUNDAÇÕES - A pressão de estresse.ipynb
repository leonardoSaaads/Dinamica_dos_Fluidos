# A pressão de estresse.

A fim de dar uma expressão para $[\sigma]$, devemos primeiro considerar o caso de um fluido homogêneo e isotrópico em equilíbrio termodinâmico. A isotropia do fluido e o fato de que o tensor de tensão depende apenas das propriedades locais do fluido exige que os valores próprios de $[\sigma]$ (que é sempre diagonalizável porque de sua simetria) são idênticos nas três direções do espaço. Portanto, nós pode escrever:

$$ \sigma_{ij} = -P \delta_{ij} $$

Onde P é uma função escalar que iremos identificar para a pressão. Alguém pode se perguntar se tal definição dá a mesma função a que estamos acostumados em Termodinâmica, nomeadamente a variável intensiva associada ao volume. Para verifique este ponto, só precisamos considerar a equação da energia interna desprezando $F$ e $\mathcal{Q}$. A variação da energia interna ıe para um pequeno volume $\triangle V$ durante $\delta t$ é:

![](https://raw.githubusercontent.com/leonardoSaaads/Dinamica_dos_Fluidos/main/Imagens/imagem_livro_2.png)

$$ \triangle p \delta t = \sigma_{ij} \partial_i v_j \triangle \delta t = -P \nabla \cdot v \triangle V \delta t = -P \delta (\triangle V)$$


Uma vez que $\rho V$ é apenas a massa $\triangle m$ do pequeno volume, $ \frac{\delta(\triangle V)}{\triangle m}$ é a variação do volume específico (o volume ocupado por uma unidade massa). Daí a relação anterior, tomada no limite $\triangle V \to 0$, nos leva a relação diferencial $-P = \frac{\partial e}{\partial v}$ onde v é o volume específico. Esta expressão é conhecida da Termodinâmica.


também descobrimos a força volumétrica associada à pressão. Tomando a divergência de $[\sigma]$:

$$ f_i = \partial_j \sigma_{ij} = - \delta_{ij} \partial_j P = - \partial_i P = -(\nabla P)_i$$


Esta expressão mostra que a força de pressão volumétrica é o oposto da pressão gradiente: as forças de pressão empurram os elementos de fluido em direção às regiões de baixa pressão como esperado. A unidade de pressão do Sistema Internacional é o Pascal, mas muitos outros ainda estão em uso.
