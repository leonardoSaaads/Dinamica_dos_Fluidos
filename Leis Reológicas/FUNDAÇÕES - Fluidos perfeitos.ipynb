# FUNDAÇÕES - FLUIDOS

## Fluidos perfeitos

A discussão anterior nos leva ao caso de um fluido ideal cujo tensor de estresse
seria composto apenas pelo termo de pressão. Esse fluido é chamado de fluido perfeito
ou um fluido invíscido. Veremos a seguir que é na verdade uma idealização conveniente
de fluidos reais.

Se escrevermos a equação de momentum para tal fluido, ele será:

$$ \rho \frac{Dv}{Dt} = -\nabla P$$

que é conhecida como equação de Euler. Aqui, todas as forças volumétricas extras, que são dependente do problema, foram removidas. Os fluidos perfeitos também são dotados da propriedade de não permitirem qualquer fluxo de calor. Conseqüentemente, $F$ está desaparecendo. Se nenhuma fonte de calor estiver presente $(\mathcal{Q}= 0)$, a equação de energia interna agora é:

$$ \rho \frac{De}{Dt} = -P \nabla \cdot v$$

e a entropia é:

$$ \frac{Ds}{Dt} = 0$$

Esta última relação nos mostra que a entropia das partículas de fluido em um fluido perfeito permanece constante. Encontramos aqui a primeira lei de conservação específica para fluidos perfeitos.

REFERÊNCIA: RIEUTORD, Michel. Fluid Dynamics An Introduction. Masson, France: Springer, 1997.
