# FUNDAÇÕES - FLUIDOS

## Fluidos Newtonianos

### Viscosidade

Todo mundo já experimentou o fluxo lento do mel em comparação com a água. Os dois
fluidos sendo de densidade semelhante, a atração da gravidade sobre as partículas de fluido é semelhante e
portanto, os comportamentos de fluxo muito diferentes que eles mostram devem estar associados
alguma propriedade intrínseca. Essa propriedade está manifestamente relacionada à capacidade do fluido
partículas deslizarem umas sobre as outras. Partículas de água deslizam com muito mais facilidade do que mel
uns! Esta característica do fluido é comumente chamada de viscosidade. Fisicamente, vemos
que esta propriedade caracteriza as interações entre elementos fluidos ou, em outras
palavras, forças de contato. Como tal, são forças superficiais e, portanto, devem ser incluídas
no tensor de tensão, adicionado à pressão. Portanto, os componentes do tensor de tensão
deve ler:

$$ \sigma_{ij} = -P \delta_{ij} + \sigma_{ij}^{viscosity}$$

A fim de descobrir a expressão deste novo termo, devemos considerar um fluido em
equilíbrio (mecânico e termodinâmico). Nós perturbamos este equilíbrio
para que o fluido se mova. Assumimos que a perturbação é pequena o suficiente para que
as partículas de fluido permanecem perto do equilíbrio termodinâmico. Precisamos agora
para descobrir qual quantidade é adequada para medir o desequilíbrio. Inicialmente
de relance, pode-se dizer que a velocidade é a quantidade adequada. Este não é o caso
(infelizmente!) porque notamos que se o campo de velocidade for uniforme, então um simples
mudança do quadro de referência fará com que ele desapareça e, portanto, os elementos fluidos são
ainda em equilíbrio. Portanto, a não uniformidade do campo de velocidade é essencial. Nós podemos
portanto, pense na derivada do campo de velocidade, ou seja, $\partial_i v_j$. Este ainda não é o
boa quantidade. Na verdade, já vimos que o tensor gradiente de velocidade
é composto por duas partes, as simétricas e as antissimétricas, que descrevem um
evolução muito diferente das partículas de fluido. A parte anti-simétrica descreve o
rotação local de partículas de fluido: pode ser zerada pelo uso de uma moldura rotativa. Nós somos
assim, fica com a parte simétrica do tensor de gradiente de velocidade, o chamado tensor de taxa de deformação, que não pode ser anulado por nenhuma mudança de quadro. A taxa de tensão
assim, aparece como a medida mais simples do desequilíbrio mecânico do fluido em
o nível de uma partícula de fluido. Portanto, escrevemos sobre um quadro rotativo.

$$ \sigma_{ij} \equiv f_{ij}(s_{kl})$$

As seis funções $f_{ij}$ são desconhecidas e dependem, a priori, das propriedades locais de
o fluido. Uma maneira de simplificar o problema é usar nossa suposição de que o fluido é
perto do equilíbrio. Nesse caso, a taxa de deformação é pequena e as funções podem
ser expandido em torno de zero:

$$ \sigma_{ij} = f_{ij}(0) + L_{ijkl}(s_{kl}) + \cdots $$

onde

$$ L_{ijkl} = \left(\frac{\partial f_{ij}}{\partial s_{kl}} \right)$$

$f_{ij}(0)$ é o valor da tensão quando a taxa de deformação é zero. É obviamente igual a $-P\delta_{ij}$ ou seja, ao estresse no equilíbrio, que vem apenas da pressão. O tensor $L$ de classificação 4 também caracteriza as propriedades de equilíbrio do fluido. Como antes de assumirmos que o fluido é isotrópico. Uma vez que L é simétrico em relação a permutação de i e j e k e l, o único tensor de classificação 4 compartilhando essas propriedades tem a forma:

$$ L_{ijkl} = a(\delta_{ik}\delta_{jl} + \delta_{jk}\delta_{il}) + b\delta_{ij}\delta_{kl}$$

onde a e b são dois coeficientes escalares específicos para o fluido. Assim encontramos

$$ \sigma_{ij} = -P\delta_{ij} + a(\partial_{i}v_j + \partial_jv_i) + b(\partial_kv_k)\delta_{ij} $$

ou

$$ \sigma_{ij} = -P\delta_{ij} + \mu \left(\partial_{i}v_j + \partial_jv_i - \frac{2}{3}(\partial_kv_k)\delta_{ij}\right) + \zeta(\partial_kv_k)\delta_{ij} $$

onde definimos $a = \mu$ e $b= \zeta - 2/3\mu$. $\mu$ é a chamada de viscosidade de cisalhamento dinâmico
enquanto $\zeta$ é a viscosidade do volume dinâmico ou viscosidade aparente (às vezes também chamada de
segunda viscosidade). Esses dois coeficientes são expressos em pascal segundo (Pa s) também
chamado de poiseuille. Note que

$$ Tr\left( \partial_iv_j + \partial_jv_i - \frac{2}{3}(\partial_kv_k)\delta_{ij}\right) = 0 $$

de modo que é natural introduzir

$$ c_{ij} =  \partial_iv_j + \partial_jv_i - \frac{2}{3}(\partial_kv_k)\delta_{ij} $$

Estes são os componentes do tensor de cisalhamento $[c]$ que corresponde a uma deformação sem mudança de volume.
Todos os fluidos que verificam a lei reológica são chamados de fluidos Newtonianos. Apesar das severas simplificações que têm sido feitas, esta constitutiva relação é verificada por um grande número de gases ou líquidos. Isso vem do fato que depende apenas de propriedades muito gerais do fluido em equilíbrio, a saber, isotropia e proximidade ao equilíbrio termodinâmico local. Também devemos, muitas vezes, usar outro coeficiente de viscosidade, a saber, o cinemático
viscosidade definida como


$$ \nu = \frac{\mu}{\rho}$$

Que é expressa em $m^2 s^{-1}$

### A Equação Momentum e a Equação de Navier-Stokes
A expressão da lei reológica para fluidos newtonianos nos permite dar uma forma explícita da equação de momento para esses fluidos.

$$ \rho \frac{Dv}{Dt} = -\nabla P + \nabla(v \cdot \nabla \mu) + \nabla \times (v \times \nabla\mu) + \mu \Delta v - v\Delta\mu + \nabla ((\mu/3 + \zeta)\nabla\cdot v) + f$$

Esta é a equação de Navier-Stokes. Esta equação simplifica muito para um fluido com densidade constante e viscosidade constante. Obtém-se então a chamada equação de Navier:

$$ \rho \frac{Dv}{Dt} = -\nabla P + \mu \Delta v$$

onde descartamos as forças externas do corpo $f$. Esta equação é usada quando a densidade
variações são insignificantes ou, se não forem, como um primeiro passo antes de atacar o
complicações devido à compressibilidade.
