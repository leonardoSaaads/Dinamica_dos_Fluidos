# FUNDAMENTOS - FLUIDOS

## Conservação da massa

Consideramos um volume fixo de fluido $V$, cuja massa é

$$\displaystyle M =  \int_V \rho dV $$

Sua variação com o tempo é dada pela densidade do fluxo de massa $\rho v$ cruzando a superfície $S$ limitando pelo volume $V$. Seja $dS$ o elemento de superfície orientado pelo versor normal externo $n$ para que $dS = n \cdot dS $. Por isso

$$ \displaystyle \frac{dM}{dt} = - \int_S \rho v \cdot dS \iff \int_V \frac{\partial \rho}{\partial t}dV = - \int_V \nabla \cdot \rho v dV \iff \int_V \left( \frac{\partial \rho}{\partial t} + \nabla \cdot \rho v \right)dV = 0 $$

Observe que o sinal de menos vem da orientação da superfície $S$ quando $v$ é paralelo a $dS$, a massa $M$ diminui. A última equação sendo verdadeira para qualquer volume, o integrando deve estar desaparecendo e nós temos

$$ \frac{\partial \rho}{\partial t} + \nabla \cdot \rho v =  0$$

que expressa localmente a conservação da massa. Esta equação é frequentemente referida como a equação de continuidade. Pode-se também escrevê-lo usando o material derivado de, nomeadamente

$$ \frac{\partial \rho}{\partial t} =  - \rho \nabla \cdot v $$

O que mostra que a densidade de um elemento fluido varia por causa de sua variação de volume expresso por $\nabla \cdot v$, pois sua massa é constante. É esperado, quando $\rho =Const$, todas essas equações levam de volta a equação acima, ou seja, o
caso de fluidos incompressíveis.

### Derivada Material com Conservação da Massa.

Na maioria dos casos, as quantidades físicas, como energia, momento, não estão ligadas ao volume dos elementos, mas à sua massa $dm = \rho dV$. Isso implica que ao escrever o equilíbrio entre perdas e ganhos para um volume fixo, sempre enfrentamos integrais levando em consideração o fluxo da quantidade transportada pelo fluxo de massa através (S). Em geral, se $\phi$ for essa quantidade física e $S_{\phi}$ suas fontes volumétricas, temos

<h4 align="center"> variações de $\phi$ em V = $\phi$ transportado por v sobre S + fontes de $\phi$ </h4> 

Matematicamente, essa afirmação é equivalente a:

$$ \displaystyle \frac{d}{dt} \int_V \rho \phi dV = - \int_S \rho \phi v \cdot dS + \int_V S_{\phi} dV $$

Note que

$$ \displaystyle \int_V \frac{\partial \rho \phi }{\partial t}dV = - \int_V \nabla \cdot (\rho \phi v ) dV + \int_V S_{\phi} dV  \iff \int_V \left(\phi\frac{\partial \rho}{\partial t} + \rho \frac{\partial \phi}{\partial t} \right)dV = - \int_V (\phi \nabla \cdot \rho v + \rho v \cdot \nabla \phi) dS + \int_V S_{\phi} dV$$

Ao usar a equação de continuidade, descobrimos que

$$ \displaystyle \int_V \rho \frac{\partial \phi }{\partial t}dV + \int_V \rho v \cdot \nabla \phi dV = \int_V S_{\phi} dV$$

ou ainda


$$ \displaystyle \int_V \rho \frac{D \phi}{Dt}dV = \int_V S_{\rho} dV $$

e essa relação é valida para qualquer fluido de volume V:

$$ \rho \frac{D \phi}{Dt} = S_{\rho} $$

Veremos que as equações de momento, energia ou entropia compartilham esse tipo de estrutura. $\rho$ é então um campo de velocidade (o momento por unidade de massa), a energia interna ou a entropia por unidade de massa.

Vamos agora rederir usando um volume $ V (t) $ anexado ao fluido. De sua definição, este volume contém as mesmas partículas de fluido em qualquer momento. Então,

<h4 align="center"> variações de $\phi$ em V(t) = fonte de $\phi$ </h4> 

Matematicamente, essa afirmação é equivalente a:

$$ \displaystyle \frac{d}{dt} \int_{V(t)} \rho \phi dV = \int_{V(t)} S_{\phi} dV $$

E então:

$$ \displaystyle \begin{align*} \frac{d}{dt} \int_{V(t)} \rho \phi dV & = \int_{V(t)} \left( \frac{D(\rho \phi)}{Dt} + \rho \phi \nabla \cdot v \right)dV \\ \, & = \int_{V(t)} \left[\phi \left( \frac{D(\rho)}{Dt} + \rho \nabla \cdot v \right) + \rho \frac{D(\rho)}{Dt}\right]dV  \\ \, & = \int_{V(t)} \rho \frac{D \phi}{Dt} dV \end{align*} $$

Nós imediatamente encontramos a relação. Esta derivação é muito mais rápida que a anterior e vamos preferir no decorrer do repositório.
