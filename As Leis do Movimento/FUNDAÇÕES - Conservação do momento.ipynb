# FUNDAÇÕES - FLUIDOS

## Conservação do momento

Antes de expressar a conservação do momento, precisamos precisar a maneira como o fluido elementos interagem. As forças de contato são específicas da mecânica dos meios contínuos. Sua existência mostra novamente o fato de que as partículas de fluido não são massas pontuais, mas pequenos volumes com uma superfície na qual as forças de contato podem ser exercidas. Vamos considerar uma superfície elementar $dS$ na qual uma força elementar $df$ é aplicada. Estes dois vetores estão relacionados de uma maneira muito geral por

$$ df = [\sigma] dS $$

ou em índices

$$ df_i = \sigma_{ij} dS_j$$

Definimos assim o tensor de tensão $[\sigma]$ e ao mesmo tempo o de tensão $T = [\sigma]n$ aplicada em um determinado ponto da superfície cuja normal é $n$. Assim definido, o estresse é uma força por unidade de superfície. Em seguida, faz-se a hipótese de que $[\sigma]$ depende apenas do local propriedades do fluxo.

Vamos mostrar agora que $[\sigma]$ é um tensor simétrico. Se $S$ é uma superfície cobrindo algum volume $(V)$, notamos que a resultante das forças de estresse em $S$ também pode ser escrita como o resultante de forças volumétricas desde

$$ \displaystyle \int_{S} \sigma_{ij} dS_j = \int_{V} \partial_j \sigma_{ij}dV$$

Portanto, sempre se pode associar uma força volumétrica a uma tensão superficial. Se considerarmos um elemento fluido, a identidade acima apenas diz que a resultante das forças de contato em um partícula de fluido é igual à divergência do tensor de tensão. Vamos agora considerar o torque exercido pelas forças de estresse no volume. Nós temos

$$ \displaystyle m_i = \int_S (r \times df)_i = \int_S (r \times [\sigma] dS)_i $$

$$ \displaystyle m_i = \int_S \epsilon_{ijk}x_j \sigma_{kl} dS_l = \int_V \partial_l (\epsilon_{ijk}x_j \sigma_{kl})dV  $$

aqui nós introduzimos a divergência vetorial Div de um tensor simétrico que é:

$$ (Div[\sigma])_i = \partial_j \sigma_{ij}= \partial_j \sigma_{ji}$$

A igualdade de expressões implica que

$$ \partial_l(\epsilon_{ijk}x_j \sigma_{kl}) = \epsilon_{ijk}x_j \partial_l\sigma_{kl} \iff \epsilon_{ijk}\delta_{lj}\sigma_{kl} + \epsilon_{ijk}x_j \partial_l\sigma_{kl} = \epsilon_{ijk}x_j x_j \partial_l\sigma_{kl} \iff \epsilon_{ijk}\sigma_{kl} = 0 $$

o que equivale a dizer que o tensor de tensão é simétrico. Então:

$$ \sigma_{ij}  = \sigma_{ji} $$

Esta equação assume implicitamente que o fluido não contém nenhuma densidade de torque

### A Equação do Momento

Estamos agora em uma posição de usar a conservação do momento para algum volume de fluido. A variação do momento de um volume de fluido, portanto, é

$$ \displaystyle \frac{d}{dt} \int_V \rho v dV = \int_V fdV + \int_S [\sigma]dS $$

Usando-se o teorma da divergência, achamos:

$$ \displaystyle \int_V \frac{Dv}{Dt}\rho dV = \int_V fdV + \int_V Div[\sigma]dV $$

ou

$$ \displaystyle \frac{Dv}{Dt}\rho = f +Div[\sigma] $$

Esta equação é apenas a segunda lei de Newton aplicada a um elemento fluido de unidade de volume. Notamos de passagem que a expressão de aceleração é

$$ a = \frac{Dv}{Dt} = \frac{\partial v}{\partial t} + (v \ \cdot \nabla) v $$

nada mais é do que a derivada material da velocidade. Esta expressão da aceleração pode ser obtida de uma maneira mais intuitiva, considerando um fluido partícula cuja trajetória é $(x(t), y(t), z(t))$. Sua velocidade no tempo $t$ onde a partícula
está em $x, y, z$ é apenas a velocidade do fluido $v(x, y, z, t)$. A velocidade ao longo da trajetória é, portanto, $v(x(t), y(t), z(t), t)$ enquanto sua aceleração, também ao longo da trajetória, é

$$ \frac{dv}{dt} = \frac{\partial v}{\partial x} \frac{dx}{dt} + \frac{\partial v}{\partial y} \frac{dy}{dt} + \frac{\partial v}{\partial y} \frac{dy}{dt} + \frac{\partial v}{\partial t}$$

Sabendo-se que $\left( \frac{dx}{dt}, \frac{dy}{dt}, \frac{dz}{dt}\right) = v$, podemos notar que:

$$ \frac{dv}{dt} = \frac{\partial v}{\partial t} + (v \cdot \nabla)v = \frac{Dv}{Dt}$$

o que mostra que a derivada material da velocidade é de fato a aceleração de a partícula de fluido. O termo $(v \cdot \nabla)v$ também pode ser escrito como $(v \cdot \nabla)v = (\nabla \times v) \times v + \nabla \frac{1}{2}v^2$, que nos dá uma outra expressão para a aceleração:

$$ \frac{Dv}{Dt} = \frac{\partial v}{\partial t} + (\nabla \times v) \times + \nabla \frac{1}{2} v^2 $$

Agora, vemos que a aceleração de um elemento fluido é controlado por duas forças volumétricas, a $f$ e $Div[\sigma]$. A força volumica $f$ é específica para os problemas em mãos. Essa força pode ser a força gravitacional (geralmente decorrente de flutuabilidade), a força de Laplace se o fluido é eletricamente condutor ou uma força inercial como a de Coriolis. Div $[\sigma]$ é a força volumétrica devido a tensões. Ao contrário de $f$, está sempre lá e representa as forças de contato entre os elementos fluidos. Depende da natureza do fluido.

___

Em suma, a variação do momento se comporá de duas partes, o momento no interior do volume de controle, e o impulso de atravessar a superfície. Este conceito pode ser escrito como:
