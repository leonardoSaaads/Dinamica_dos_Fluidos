# FUNDAÇÕES - FLUIDOS

## Conservação da energia

A equação que traduz a conservação de energia expressa o primeiro princípio de termodinâmica com um elemento fluido. O balanço de energia é dado:

$$ \displaystyle \frac{d}{dt} \int_V \rho \left(\frac{v^2}{2} + e \right)dV = \int_V f \cdot v dV + \int_S v_i \sigma_{ij}dS_j - \int_S F \cdot dS + \int_V \mathcal{Q} dV$$

onde $e$ denota a energia interna específica, $F$ é a (densidade superficial) do fluxo de calor e $\mathcal{Q}$ a potência das fontes de calor locais. Essas fontes podem vir de produtos químicos reações (queima), de reações nucleares (na parte central das estrelas) ou de um transição de fase (liberação de calor latente na condensação de vapor d'água, por exemplo). Usando palavras, a última equação seria:

**A variação de energia (cinética e interna) por unidade de tempo = o poder das forças volumétricas + o poder das tensões + o fluxo de calor através de S + o poder das fontes de calor locais**


$$ \displaystyle  \int_V \rho \frac{D \left(\frac{v^2}{2} + e \right) }{Dt}dV = - \int_V \nabla \cdot F dV + \int_V f \cdot v dV  + \int_V \partial_j(v_i \sigma_{ij})dV + \int_V \mathcal{Q} dV$$

sendo o volume qualquer volume, a equação é válida localmente, portanto:

$$ \displaystyle \rho \frac{D \left(\frac{v^2}{2} + e \right) }{Dt} = - \nabla \cdot F + f \cdot v + \partial_j(v_i \sigma_{ij}) + \mathcal{Q}$$

### A Equação para a Energia Interna


A equação de energia anterior combina as energias internas e cinéticas. No entanto, a evolução da energia cinética é governada pela única equação de momentum. Na verdade, tomando o produto escalar $\rho \frac{Dv}{Dt} = Div[\sigma] + f$ com $v$, encontramos

$$ \displaystyle \rho \frac{D \left(\frac{v^2}{2}\right)}{Dt} = f \cdot v + v_i \partial_j \sigma_{ij} $$

que é apenas a expressão do teorema da energia cinética aplicada a um elemento fluido: a mudança da energia cinética de um elemento fluido vem do trabalho de aplicação forças. Subtraindo a evolução da energia cinética, encontramos a evolução de energia interna, isto é:

$$ \displaystyle \rho \frac{De}{Dt} = - \nabla \cdot F + \sigma_{ij} \partial_j v_i + \mathcal{Q}$$

que expressa localmente o primeiro princípio da termodinâmica: a variação de a energia interna de uma partícula de fluido é igual ao calor recebido $(-\nabla \cdot F + \mathcal{Q})$ mais o trabalho das tensões $\sigma_{ij} \partial_j v_i$. Devemos notar que este trabalho depende exclusivamente do tensor de taxa de deformação local $[s]$ visto que $[\sigma]$ é simétrico, $\sigma_{ij} \partial_j v_i = \sigma_{ij}s_{ij}$.

### A Equação para a Entropia

Em vez de usar a energia interna para expressar a conservação de energia, muitas vezes é útil para escolher a entropia. Isso é facilmente derivado das igualdades relacionadas ao várias quantidades termodinâmicas

$$ de = TdS - PdV = Tds + Pd\rho/\rho^2 $$

Essas equações ligam as derivadas totais e, portanto, se aplicam a todas as derivadas parciais com respeito ao tempo e espaço. Portanto, eles podem ser combinados para produzir uma relação entre derivadas de materiais:

$$\frac{De}{Dt} = T\frac{Ds}{Dt} + \frac{P}{\rho^2}\frac{D\rho}{Dt} $$

Combinado com a conservação da massa, isso nos dá:

$$ T\rho \frac{Ds}{Dt} = -\nabla \cdot F + \sigma_{ij}\partial_j v_i + P \nabla \cdot v + \mathcal{Q} $$

Esta equação expressa que a variação da entropia dos elementos fluidos é o resultado de fontes de calor dentro dos elementos mais o fluxo de calor vindo do vizinho elementos, além de alguma geração de calor devido a tensões (viscosas.
